### Exercício 1: Cálculo de Área de um Círculo

Escreva um programa que calcula a área de um círculo. O usuário deve fornecer o raio e o programa deve retornar a área.

```python
# Importa o módulo math para acessar constantes e funções matemáticas, como o valor de pi
import math

def area_circulo(raio):

    # Calcula a área do círculo usando a fórmula área = pi * raio^2
    # math.pi fornece o valor de pi
    return math.pi * raio ** 2

# Pede ao usuário para inserir o valor do raio e converte para número ponto flutuante
raio = float(input("Digite o raio do círculo: "))

# Imprime a área do círculo chamando a função acima com o raio fornecido.
print("A área do círculo é:", area_circulo(raio))
```
